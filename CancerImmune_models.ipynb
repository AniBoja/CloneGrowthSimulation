```julia
using Distributions, DataFrames,  Plots, ODE
```

```julia
# population grows according to birthrate and cell number
# dN/dt = b*N
# population gets killed with function according to neoep burden, cell number, additional immune par
```

```julia
f(s, rMax, kLim, a) = rMax./(1+exp.(-a*(s-kLim)))
g(e, rMax, kLim, a) = rMax./(1+exp.(-a*(e-kLim)))
plot(0:0.05:1000, f(0:0.05:1000, 4, 200, 0.03), linewidth=3)
plot(0:0.05:400, g(0:0.05:400, 4, 200, 0.015), linewidth=3)
```

```julia
h(s, e) = f(s, 2, 300, 0.03).*g(e, 2, 200, 0.015)
ts = 0:0.5:1000
te = 0:0.5:500
xE = plot(ts, [h(ts, 10), h(ts, 100), h(ts, 200), h(ts, 300), h(ts, 400)])
#xS = plot(te, [h(100, te), h(200, te), h(300, te), h(500, te)])
```

```julia
function myode(u,p,t)
    b=1.05
    e = 2
    du = b*u - e/100*u 
    return(du)
end

u0= 1.0
tspan = (0.0,5.0)
prob = ODEProblem(myode,u0,tspan)
sol = solve(prob,ode45(),reltol=1e-8,abstol=1e-8)

plot(sol,linewidth=5,title="Solution to the linear ODE with a thick line",
     xaxis="Time (t)",yaxis="u(t) (in μm)",label="My Thick Line!") # legend=false
```

```julia
function myode(u,p,t)
    return(1.01*u)
end
u0=1/2
tspan = (0.0,1.0)
prob = ODEProblem(myode,u0,tspan)
sol = solve(prob,ode45(),reltol=1e-8,abstol=1e-8)

plot(sol,linewidth=5,title="Solution to the linear ODE with a thick line",
     xaxis="Time (t)",yaxis="u(t) (in μm)",label="My Thick Line!") # legend=false
plot!(sol.t, t->0.5*exp(1.01t),lw=3,ls=:dash,label="True Solution!")
```
